```python
urls = ['https://inshorts.com/en/read/sports',
        'https://inshorts.com/en/read/world',
        'https://inshorts.com/en/read/politics']

def build_dataset(urls):        
  news_data = []
  for url in urls:
    news_category = url.split('/')[-1]
    data = requests.get(url)
    soup = BeautifulSoup(data.content)

    news_articles =[{'news_headline':headline.find('span',attrs={"itemprop":"headline"}).string,
                     'news_article':article.find('div',attrs={"itemprop":"articleBody"}).string,
                     'news_category':news_category}
                    
                    for headline,article in zip(soup.find_all('div',class_=["news-card-title news-right-box"]),soup.find_all('div',class_=["news-card-content news-right-box"]))]
    news_articles = news_articles[0:1000]
    news_data.extend(news_articles)
    df = pd.DataFrame(news_data)
    df1 = df[["news_headline","news_article","news_category"]]
    return df1            
    # news_articles = news_articles[0:20]
    # news_data.extend(news_articles)

# df = pd.DataFrame(news_articles)
# df1 = df[["news_headline","news_article","news_category"]]
# return df1                                             
```

```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/Sumitdas009/ML-MAJOR-MAR/main/news.csv')
```

```python
import nltk
nltk.download('stopwords')

stopword_list = nltk.corpus.stopwords.words('english')
stopword_list.remove('no')
stopword_list.remove('not')
print(stopword_list)
len(stopword_list)
```

```python
#Remove HTML_tag(text)
def html_tag(text):
   soup  = BeautifulSoup(text,"html.parser")
   new_text = soup.get_text()
   return new_text

html_tag('<html><h2>Some Imp Info</h2></html>')   
```

```python
#Expand Contractions
!pip install contractions
import contractions
def con(text):
  expand = contractions.fix(text)
  return expand

con("Y'all")
```

```python
#Special characters
import re
def remove_sp(text):
  pattern = r'[^A-Za-z0-9]'
  text =re.sub(pattern,' ',text)
  return text

remove_sp("+-*skdm jid")  
```

```python
from nltk.tokenize.toktok import ToktokTokenizer
tokenizer = ToktokTokenizer()
```

```python
def remove_stopwords(text):
  tokens = tokenizer.tokenize(text)
  tokens = [token.strip() for token in tokens]
  filtered_tokens = [token for token in tokens if token not in stopword_list]
  filtered_text = ' '.join(filtered_tokens)
  return filtered_text
```

```python
remove_stopwords("The, and , if are not even stopwords")
```

```python
df.news_headline = df.news_headline.apply(lambda x:x.lower())
df.news_article = df.news_article.apply(lambda x:x.lower())

df.news_headline = df.news_headline.apply(html_tag)
df.news_article = df.news_article.apply(html_tag)

df.news_headline = df.news_headline.apply(con)
df.news_article = df.news_article.apply(con)

df.news_headline = df.news_headline.apply(remove_sp)
df.news_article = df.news_article.apply(remove_sp)

df.news_headline = df.news_headline.apply(remove_stopwords)
df.news_article = df.news_article.apply(remove_stopwords)

df.head()
```

```python
df['compound'] = df['news_article'].apply(lambda x: vs.polarity_scores(x)['compound'])
```

```python
df.head()
```

```python
def predict(comp):
  comp=float(comp)
  if (comp>=0):
    return 'positive'
  else:
    return 'negative'
df['type_pred'] = df['compound'].apply(predict)
df.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = df.iloc[:,0].values.astype('U')
y = df.iloc[:,4].values.astype('U')
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=420,test_size=0.3)
x_train
```

```python
y_train
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import Pipeline

model=Pipeline([('TFIDF',TfidfVectorizer()),('model',MultinomialNB())])
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
accuracy_score(y_pred,y_test)
```

```python
confusion_matrix(y_pred,y_test)
```

```python
print(classification_report(y_pred,y_test))
```
